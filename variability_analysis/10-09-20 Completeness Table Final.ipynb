```python
def compute_completeness(df):
    '''
    Computes completeness of data for a location
    '''
    total_hours = 4368
    actual_hours = df.shape[0]
    return actual_hours, round(actual_hours/total_hours*100, 2)
```

```python
def pm_min(df):
    '''Calculates the minimum PM 2.5 value of a location'''
    return round(df['PM 2.5'].min(), 1)
```

```python
def pm_max(df):
    '''Calculates the minimum PM 2.5 value of a location'''
    return round(df['PM 2.5'].max(), 1)
```

```python
def pm_mean(df):
    '''Calculates the average PM 2.5 value of a location'''
    return round(df['PM 2.5'].mean(), 1)
```

```python
def pm_median(df):
    '''Calculates the median PM 2.5 value of a location'''
    return round(df['PM 2.5'].median(), 1)
```

```python
def pm_percentile_25(df):
    '''Calculates the 25th Percentile PM 2.5 value of a location'''
    return round(df['PM 2.5'].quantile(0.25), 1) 
```

```python
def pm_percentile_75(df):
    '''Calculates the 75th Percentile PM 2.5 value of a location'''
    return round(df['PM 2.5'].quantile(0.75), 1) 
```

```python
def get_delta(start, end):
    '''Calculates difference in hours between two datetimes'''
    return abs(int((end-start).total_seconds())) / 3600
```

```python
def resample_time(df): #df should have a datetime index
    '''
    Returns dataframe with hourly mean values based on pre-defined criteria
    '''
    from datetime import timedelta
    new_df = pd.DataFrame(columns =['created_at', 'PM 2.5'])
    start = df.index.min().replace(microsecond=0, second=0, minute=0)
    end = df.index.max().replace(microsecond=0, second=0, minute=0)+ timedelta(hours=1)
    for h in range(int(get_delta(start, end))):
        time = start + timedelta(0, h*3600)
        time_df = df.loc[time:time+timedelta(hours=1)]
        if time_df.shape[0]>=30:
            mean_pm = round(time_df['PM 2.5'].mean(), 2)
            new_row = {'created_at':time, 'PM 2.5':mean_pm}
            new_df = new_df.append(new_row, ignore_index=True)
        else:
            pass
    return new_df
```

```python
def preprocessing(df):
    '''
    Preprocesses data for a location
    '''
    df['created_at'] = pd.to_datetime(df['created_at'])
    df = df.sort_values(by='created_at',ascending=True)
    df = df.set_index('created_at')
    df = df.rename(columns={'field1':'PM 2.5'})
    df['PM 2.5'].fillna(df['field3'])
    df.drop(['field3'], axis=1, inplace=True)
    df['PM 2.5'] = pd.to_numeric(df['PM 2.5'],errors='coerce')
    hourly_df = resample_time(df)
    hourly_df.dropna(inplace=True)
    return hourly_df
    #df.dropna(inplace=True)
    #return df
```

```python
def overall_func(myArray):
    columns = ['Parish', 'Latitude', 'Longitude', 'Minimum', 'Maximum', 'Mean PM 2.5', 'Median PM 2.5', '25th Percentile', '75th Percentile', 'No. of hourly measurements', '6-Month Completeness (%)']
    final_df= pd.DataFrame(columns=columns)
    #df = pd.read_csv('./data/'+location['name']+'-'+location['location']+'.csv', 
     #                    usecols = ['created_at', 'field1', 'field3'],
      #                  parse_dates =['created_at'])
    #preprocessed_df = preprocessing(df)
    for location in myArray:
        df = pd.read_csv('./data/'+location['name']+'-'+location['location']+'.csv', 
                         usecols = ['created_at', 'field1', 'field3'],
                        parse_dates =['created_at'])
        preprocessed_df = preprocessing(df)
        hourly_measurements, completeness = compute_completeness(preprocessed_df)
        minimum = pm_min(preprocessed_df)
        maximum = pm_max(preprocessed_df)
        mean = pm_mean(preprocessed_df)
        median = pm_median(preprocessed_df)
        percentile_25 = pm_percentile_25(preprocessed_df)
        percentile_75 = pm_percentile_75(preprocessed_df)
        df2 = pd.DataFrame([[location['location'], location['lat'], location['long'], minimum, maximum, mean, median, percentile_25, percentile_75, hourly_measurements, completeness ]], columns=columns)
        final_df = pd.concat([df2, final_df])
        print(location['location']+' done')
    final_df.to_csv('data/completeness_table_10_09_2020.csv', index=False)
    return final_df
```
