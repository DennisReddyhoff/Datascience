```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
import requests
from datetime import datetime
%matplotlib inline
```

```python
def preprocessing(df): #hasn't yet been tested
    '''
    cleaning/preprocessing data in dataframe
    '''
    df['created_at'] = pd.to_datetime(df['created_at'])
    df = df.sort_values(by='created_at',ascending=False)
    #df = df.reset_index(drop=True)
    df = df.set_index('created_at')
    #df = df.rename(columns={'field1':'PM 2.5', 'field2':'PM 10'})
    df = df.rename(columns={'field1':'PM 2.5'})
    df['PM 2.5'].fillna(df['field3'])
    #df['PM 10'].fillna(df['field4'])
    #dropping unwanted columns
    df.drop(['entry_id','field2','field3', 'field4', 'field5', 'field6', 'field7', 'field8'], axis=1, inplace=True)
    #converting datatype
    df['PM 2.5'] = pd.to_numeric(df['PM 2.5'],errors='coerce')
    #converting to hourly measurements
    hourly_df = df.resample('H').mean()
    hourly_df.dropna(inplace=True)
    return hourly_df
```

```python
def compute_completeness(df):
    total_hours = 4368
    actual_hours = df.shape[0]
    return actual_hours, round(actual_hours/total_hours*100, 2)
```

```python
def pm_min(df):
    return round(df['PM 2.5'].min(), 1)
```

```python
def pm_mean(df):
    return round(df['PM 2.5'].mean(), 1)
```

```python
def pm_median(df):
    return round(df['PM 2.5'].median(), 1)
```

```python
def pm_percentile_25(df):
    return round(df['PM 2.5'].quantile(0.25), 1) # 25th percentile
```

```python
def pm_percentile_75(df):
    return round(df['PM 2.5'].quantile(0.75), 1) # 75th percentile
```

```python
def overall_func(myArray):
    columns = ['Parish', 'Latitude', 'Longitude', 'Minimum', 'Mean PM 2.5', 'Median PM 2.5', '25th Percentile', '75th Percentile', 'No. of hourly measurements', '6-Month Completeness (%)']
    final_df= pd.DataFrame(columns=columns)
    for location in myArray:
        df = pd.read_csv('./data/'+location['name']+'-'+location['location']+'.csv')
        preprocessed_df = preprocessing(df)
        hourly_measurements, completeness = compute_completeness(preprocessed_df)
        minimum = pm_min(preprocessed_df)
        mean = pm_mean(preprocessed_df)
        median = pm_median(preprocessed_df)
        percentile_25 = pm_percentile_25(preprocessed_df)
        percentile_75 = pm_percentile_75(preprocessed_df)
        df2 = pd.DataFrame([[location['location'], location['lat'], location['long'], minimum, mean, median, percentile_25, percentile_75, hourly_measurements, completeness ]], columns=columns)
        final_df = pd.concat([df2, final_df])
        print(location['location']+' done')
    #final_df.to_csv('data/completeness_table.csv', index=False)
    return final_df
```
